```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster
```

```python
wine = pd.read_csv('/content/wine.csv')
```

```python
wine
```

```python
wine['Type'].value_counts()
```

```python
wines = wine.drop(['Type'],axis=1)
```

```python
wines.head(5)
```

```python
wines.info()
```

```python
wines.isna().sum()
```

```python
wines.describe()
```

```python
# Converting data to numpy array
wines_ary=wines.values
```

```python
wines_ary
```

```python
# Normalization the numerical data
wines1=scale(wines_ary)
wines1
```

```python
pca = PCA(n_components = 13)
pca_values = pca.fit_transform(wines1)
pca_values
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,decimals = 8)*100)
print("The amount of varience that each PCA explains is",var1)
```

```python
pca.components_
```

```python
plt.plot(var1,color='orange')
```

```python
wine_new = pd.concat([pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3']), wine['Type']], axis =1)
```

```python
wine_new
```

```python
# Visualization of PCAs
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data = wine_new)
```

```python
# create dendrogram
plt.figure(figsize = (12, 7));
dendrogram = sch.dendrogram(sch.linkage(wines1, method='complete'))
```

```python
# create clusters
hc1 = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'average')
```

```python
# save clusters for chart
y2 = hc1.fit_predict(wine)
cc = pd.DataFrame(y2,columns=['Clusters'])
cc.head(6)
```

```python
wine2 = pd.concat([wine,cc],axis=1)
wine2.head(7)
```

```python
wine2['Clusters'].value_counts()
```

```python
wcss = []
for i in range(1, 8):
  kmeans = KMeans(n_clusters=i,random_state=2)
  kmeans.fit(wine)
  wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1, 8), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Build Cluster algorithm
cc2 = KMeans(3,random_state=27)
cc2.fit(wine)
```

```python
cc2.labels_
```

```python
# Converting array dataframe
df2 = pd.DataFrame(cc2.labels_, columns = ['clusters'])
```

```python
wine3 = pd.concat([wine,df2],axis=1)
wine3
```

```python
wine3.groupby('clusters').agg(['mean'])
```

```python
model = TSNE(n_components=2)
data_tsne = model.fit_transform(wine.iloc[:,1:])
data_tsne
```

```python
# TSNE Visualization
# Plot will be stored in the default directory
cluster.tsneplot(score=data_tsne,
                 colorlist = wine['Type'],
                 legendpos = 'upper right')
```
